```python
# Task 1
# 1. Blood glucose levels for obese patients have a mean of 100 with a standard deviation of 15. A
# researcher thinks that a diet high in raw cornstarch will have a positive effect on blood glucose
# levels. A sample of 36 patients who have tried the raw cornstarch diet have a mean glucose
# level of 108. Test the hypothesis that the raw cornstarch had an effect or not.

import scipy.stats as st
import math as mt
from scipy.stats import zscore

#Hypothesis Assumptions
# 
# H0: μ= 100
# H1: μ > 100

m=100
sd=15
sample=36
samplemean=108
standarderror=0.05 #considering since we have multiple samples

# Standard Z Score = (score-mean) / (standard deviation-square root (sample))
zs=(samplemean-m)/(sd/mt.sqrt(sample))

zval=st.norm.cdf(zs)  # Getting Z Score from Z-Table

print("Z Value is : "+str(zval))

if(zval < standarderror):
    print("Computed ZVal is less than Standard Error. Null Hypothesis is rejected")
else:
    print("Computed ZVal is more than Standard Error. Alternate Hypothesis is rejected")
    
```

```python
# 2. In one state, 52% of the voters are Republicans, and 48% are Democrats. In a second state,
# 47% of the voters are Republicans, and 53% are Democrats. Suppose a simple random sample
# of 100 voters are surveyed from each state.
# What is the probability that the survey will show a greater percentage of Republican voters in
# the second state than in the first state?
import scipy.stats as st
import math as mt
from scipy.stats import zscore

# s stands for State, r stands for Republicans, d stands for democrats
s1r,s1d,s2r,s2d=52/100,48/100,47/100,53/100

# Find the mean of the difference in sample proportions
republic_mean=(s1r-s2r)
print("Mean is  : "+str(republic_mean))

n=100
# Find standard deviation difference
# Formular σd = sqrt{ [ P1(1 - P1) / n1 ] + [ P2(1 - P2) / n2 ] } 

sd = mt.sqrt(((s1r * (1-s1r)) / n) + ((s2r * (1-s2r)) / n))
print("Standard Deviation Difference is  : "+str(sd))

sdr1=s1r * (1-s1r)
print("Standard Deviation for State 1 Republicans : "+str(sdr1))

# Standard Z Score = (score-mean) / (standard deviation-square root (sample))
zval=(0-republic_mean)/(sd)

print("Z Value is : "+str(zval))

zs=st.norm.cdf(zs)  # Getting Z Score from Z-Table

print("Z Score is : "+str(zs))

if(sdr1 < republic_mean):
    print("The probability that the survey will show a greater percentage of Republican voters in the first state !!!")
else:
    print("The probability that the survey will show a greater percentage of Republican voters in the second state !!!")
```

```python
# 3. You take the SAT and score 1100. The mean score for the SAT is 1026 and the standard
# deviation is 209. How well did you score on the test compared to the average test taker?

import scipy.stats as st
import math as mt
from scipy.stats import zscore
import numpy as np

#Hypothesis Assumptions
# 
# H0: μ= average test scorer
# H1: μ > average test scorer


m=1026
sd=209
samplemean=1100
standarderror=0.05 #considering since we have multiple samples

# Standard Z Score = (score-mean) / (standard deviation-square root (sample))
zs=(samplemean-m)/(sd)

print("Z Value is : "+str(zs))

zval=st.norm.cdf(zs)

print("Z Score is : "+str(zval))

if(zval < standarderror):
    print("Alternate hypothesis rejected...I have not performed better than the average test taker.")
else:
    print("Null hypothesis rejected...I have performed better than ",np.round(zval*100,2),"% of the average test taker .")
```

```python
# Task 2:
# 1. Is gender independent of education level? A random sample of 395 people were surveyed
# and each person was asked to report the highest education level they obtained. The data that
# resulted from the survey is summarized in the following table:
# High School Bachelors Masters Ph.d. Total
# Female 60 54 46 41 201
# Male 40 44 53 57 194
# Total 100 98 99 98 395
# Question: Are gender and education level dependent at 5% level of significance? In other
# words, given the data collected above, is there a relationship between the gender of an
# individual and the level of education that they have obtained?

#Hypothesis Assumptions
# 
# H0: μ= gender of individual are independent
# H1: μ > gender of individual are dependent

# Expected Value matrix
# E(r,c) = [n(r) x c(r)] / n

female = [60, 54, 46, 41] 
male = [40, 44, 53, 57] 
print("Observed Female Values",list(female))
print("\nObserved Male Values",list(male))

f_total,m_total,p_total = 201,194,395
e_total = [100, 98, 99, 98]
significance_level = 0.05
sample_size = 4

# Calculating the expected values for obtaining the chi square
# Formula
# (column total * series total) / total samples
def calc_expected_values(comm_total_lst, seq_total, sample_len): #Calculates expected values 
    lst = []
    for val in comm_total_lst:
        lst.append(val*seq_total/sample_len)
    return lst

expected_female = calc_expected_values(e_total,f_total,p_total)
print("\nExpected Female Values",list(expected_female))
expected_male = calc_expected_values(e_total,m_total,p_total)
print("\nExpected Male Values",list(expected_male))

# Calculating Degree Of Freedom
# DOF = number of samples - 1
dof=sample_size-1
print("\nDegree Of Freedom",dof)

chi_square_observed = st.chisquare((male+female),(expected_male+expected_female)) #Calculated Chi-Square for given observed and expected values 
print("\nChi-Square (Observed) ",chi_square_observed.statistic)

chi_square = st.chi2.isf(q=significance_level, df=dof)
print("\nChi-Square Critical (Actual) ",chi_square)

if chi_square < chi_square_observed.statistic:
    print("\nNull Hypothesis Rejected.Eduation Level depends on Gender !!!")
else:
    print("\nAlternate Hypothesis Rejected.Eduation Level does not depends on Gender !!! ")
```

```python
# 3. Calculate F Test for given 10, 20, 30, 40, 50 and 5,10,15, 20, 25.
# For 10, 20, 30, 40, 50:

import scipy.stats as st
import statistics as sts

# Hypothesis Assumptions
#         H0: No difference in variances
#         HA: difference in variances.

lst_1 = [ 10, 20, 30, 40, 50] 
print("Given first data set:",lst_1)

lst_2 = [ 5, 10, 15, 20, 25] 
print("Given second data set:",lst_2)

var_lst_1 = sts.variance(lst_1)
print("\nVariance of List 1:",var_lst_1)

var_lst_2 = sts.variance(lst_2) #Variance of 5, 10, 15, 20, 25
print("Variance of List 1:",var_lst_1)

F_observed = var_lst_1/var_lst_2 
print("\nF-Test value of the given data sets (var_1/var_2):",F_observed)

dof_lst_1 = len(lst_1)-1 
print("\nDegree of freedom of 10, 20, 30, 40, 50 is:",dof_lst_1)

dof_lst_2 = len(lst_2)-1 
print("\nDegree of freedom of 5, 10, 15, 20, 25 is:",dof_lst_2)

significance_level = 0.05 

F_Critical = st.f.ppf(q=1-significance_level, dfn=dof_lst_1, dfd=dof_lst_2) # F-Table value using default alpha value = 0.05 and calculated degree of freedom
print("\nF-Critical value using default alpha (value = 0.05) is: "+str(F_Critical))

if F_Critical < F_observed:
    print("\nNull hypothesis rejected.")
else:
    print("\nActual hypothesis rejected.")
```
